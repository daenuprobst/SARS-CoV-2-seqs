```python
# Setup
import pandas as pd
```

```python
df = pd.read_csv("data/annotated_seqs.csv.gz", sep=";")
# Let's only get the spike
df = df[df.gene == "S"]
```

```python
print("total", len(df))
print(df.groupby("country").size().to_string())
```

```python
# Looks like Europe doesn't sequence or doesn't care to submit sequences to a open database.
# Looks like currently it's almost exclusively the US
df[df.country == "USA"].groupby("collection_date").size().plot()
df[df.country == "Australia"].groupby("collection_date").size().plot()
df[df.country == "Germany"].groupby("collection_date").size().plot()
df[df.country == "India"].groupby("collection_date").size().plot()
```
